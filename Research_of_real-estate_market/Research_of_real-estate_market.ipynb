### Поиск неочевидных пропусков:

```python
#Проверим столбцы с типом данных object на неочевидные пропуски

print('Уникальные значения маркера апартаменов: ')
print(df['is_apartment'].unique())
print()

print('Уникальные значения локаций: ')
print(df['locality_name'].sort_values().unique())
print()


```

В данных отсутсвуют неочевидные пропуски (строки типа "Unknown", 'None' и тп)

### Обработка пропусков:

```python
#сохраним количество строк до обработки
before = len(df)

#заполняем пропуски в высоте потолков медианными значениями
df['ceiling_height'] = df['ceiling_height'].fillna(df['ceiling_height'].median())

#заполняем пропуски в количестве этажей дома медианными значениями
df['floors_total'] = df['floors_total'].fillna(df['floors_total'].median())

#заполняем пропуски в жилой площади медианой
df['living_area'] = df['living_area'].fillna(df['living_area'].median())

#заполняем пропуски в маркере апартаментов, площади кухни, количества балконов на "0"
df['is_apartment'] = df['is_apartment'].fillna(0)
df['kitchen_area'] = df['kitchen_area'].fillna(0)
df['balcony'] = df['balcony'].fillna(0)

#заполняем пропуски в расстоянии до центра медианными значениями
df['city_centers_nearest'] = df['city_centers_nearest'].fillna(df['city_centers_nearest'].median())

#заполняем пропуски в расстоянии до парка и водоема медианами
df['parks_nearest'] = df['parks_nearest'].fillna(df['parks_nearest'].median())
df['ponds_nearest'] = df['ponds_nearest'].fillna(df['ponds_nearest'].median())
df['airports_nearest'] = df['airports_nearest'].fillna(df['airports_nearest'].median())
df['city_centers_nearest'] = df['city_centers_nearest'].fillna(df['city_centers_nearest'].median())
df['parks_around3000'] = df['parks_around3000'].fillna(df['parks_around3000'].median())
df['ponds_around3000'] = df['ponds_around3000'].fillna(df['ponds_around3000'].median())

#удаляем строки с неуказанным параметром города
df = df.dropna(axis='index', how='any', subset=['locality_name'])

#удаляем строки с неуказанным сроком
df = df.dropna(axis='index', how='any', subset=['days_exposition'])

#заполняем пропуски в сроке продажи медианой
df['days_exposition'] = df['days_exposition'].fillna(df['days_exposition'].median())

#сохраним количество строк после обработки
after = len(df)
```

### Проверка обработки пропусков:

```python
#Создадим новый столбец и поместим туда отношение цены квартиры к площади

df['sq_meter_price'] = df['last_price']/df['total_area']

#Проверим результат
df.head()
```

### Добавляем столбцы с днём недели, месяцем и годом публикации объявления

```python
# формируем нужные столбцы
df['pub_week_day'] = pd.DatetimeIndex(df['first_day_exposition']).weekday
df['pub_month'] = pd.DatetimeIndex(df['first_day_exposition']).month
df['pub_year'] = pd.DatetimeIndex(df['first_day_exposition']).year

# проверяем результаты
df.head()

```

### Добавляем информацию об этаже квартиры

```python
# создадим функцию для категоризации

def floor_categorizer(row):
    if row ['floor'] == 1: return 'первый'
    elif row['floor'] == row['floors_total'] : return 'последний'
    else : return 'другой'
    
# применим функцию ко всем строкам датафрейма
df['floor_category'] = df.apply(floor_categorizer, axis = 1)

# проверим результат

df.head(12)
```

### Добавляем информацию о жилой площади и площади кухни

```python
#создадим функцию для подсчета отношения жилой площади к общей
def live_ratio(row):
    return row['living_area']/row['total_area']

#создадим функцию для подсчета отношения площади кухни к общей
def kitchen_ratio(row):
    return row['kitchen_area']/row['total_area']

# применим функции ко всем строкам датафрейма
df['live_ratio'] = df.apply(live_ratio, axis = 1)
df['kitchen_ratio'] = df.apply(kitchen_ratio, axis = 1)

#проверим результат
df.head()
```
